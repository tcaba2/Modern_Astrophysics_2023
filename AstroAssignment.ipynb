```python
#COMPUTING MU

d=4530.8 #pc
mu=5*(np.log10(d)-1)
mu
```

```python
#COMPUTING AG (reddenning)
ebv = 0.15
ebr = 1.339*ebv
Ag = 1.89*ebr
Ag
```

```python
#FIRST CORRECTION: BY MU
plt.figure(figsize=(10,6))

fig, axes = plt.subplots(1,1)

plt.plot(df.gaiaSource_bpRp, df.gaiaSource_photGMeanMag, '+', markersize=0.5, color='red',label='NGC 6791')

plt.plot(iso5.G_BP - iso5.G_RP, iso5.G+mu+Ag, label='5000Myr',color='orange')
plt.plot(iso6.G_BP - iso6.G_RP, iso6.G+mu+Ag, label='6000Myr',color='green')
plt.plot(iso7.G_BP - iso7.G_RP, iso7.G+mu+Ag, label='7000Myr', color='black')
plt.plot(iso8.G_BP - iso8.G_RP, iso8.G+mu+Ag, label='8000Myr', color = 'blue')
plt.plot(iso9.G_BP - iso9.G_RP, iso9.G+mu+Ag, label='9000Myr', color = 'violet')

#set limits
axes.set_xlim(0,3)
axes.set_ylim(22,12)

#axes.set_title('Isochrones corrected for $\mu$')
axes.set_xlabel('Bp-Rp [mag]')
axes.set_ylabel('G [mag]')
#axes.invert_yaxis()

plt.legend(prop={'size': 10})
plt.show()
```

```python
#final corrected isochrones and data
plt.figure(figsize=(6,5))

fig, axes = plt.subplots(1,1)
plt.plot(df.gaiaSource_bpRp, df.gaiaSource_photGMeanMag, '+', markersize=0.5, color='red',label='NGC 6791')

plt.plot(iso5.G_BP - iso5.G_RP+ebr, iso5.G+mu+Ag, label='5000 Myr', color='orange', lw=1)
plt.plot(iso6.G_BP - iso6.G_RP+ebr, iso6.G+mu+Ag, label='6000 Myr', color = 'green', lw=1)
plt.plot(iso7.G_BP - iso7.G_RP+ebr, iso7.G+mu+Ag, label='7000 Myr', color='black', lw=1)
plt.plot(iso8.G_BP - iso8.G_RP+ebr, iso8.G+mu+Ag, label='8000 Myr', color = 'blue', lw=1)
plt.plot(iso9.G_BP - iso9.G_RP+ebr, iso9.G+mu+Ag, label='9000 Myr', color = 'violet', lw=1)

#axes.set_title('Corrected isochrones & NGC 6791 data')
axes.set_xlabel('Bp-Rp [mag]')
axes.set_ylabel('G [mag]')
axes.invert_yaxis()
axes.set_xlim(0.5,2.25)
axes.set_ylim(21,13)

plt.legend(prop={'size': 12}, loc='center right')
plt.show()
```
